### 11. Use NumPy to calculate the correlation coefficient between mass and orbital_period. What does this value indicate about their relationship?

```python
planets1=planets.dropna(subset=['mass','orbital_period'])
```

```python
mass=np.array(planets1['mass'])
orbital=np.array(planets1['orbital_period'])
```

```python
np.corrcoef(x=mass,y=orbital)
```
